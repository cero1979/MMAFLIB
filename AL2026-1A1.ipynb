### Â¿CÃ³mo medir la similitud entre pelÃ­culas?

AquÃ­ es donde entra el **Ã¡lgebra lineal**. Si dos pelÃ­culas tienen vectores que **apuntan en la misma direcciÃ³n**, son similares. Si apuntan en **direcciones perpendiculares** (ortogonales), son totalmente diferentes.

La **proyecciÃ³n ortogonal** nos permite medir exactamente esto: **cuÃ¡nto de un vector "va en la direcciÃ³n" de otro vector**.

> ğŸ’¡ **ConexiÃ³n con la industria:** En la prÃ¡ctica, los sistemas reales de IA usan vectores con **cientos o miles de componentes** (no solo 2). Pero el principio matemÃ¡tico es exactamente el mismo que vas a explorar aquÃ­ con 2 componentes.

---

### ğŸ“ Recordatorio matemÃ¡tico: ProyecciÃ³n Ortogonal

Dado un vector $\vec{a}$ y un vector $\vec{b}$ (con $\vec{b} \neq \vec{0}$), la **proyecciÃ³n ortogonal de $\vec{a}$ sobre $\vec{b}$** es el vector:

$$\text{proy}_{\vec{b}}(\vec{a}) = \frac{\vec{a} \cdot \vec{b}}{\vec{b} \cdot \vec{b}} \, \vec{b}$$

Donde:
- $\vec{a} \cdot \vec{b} = a_1 b_1 + a_2 b_2$ es el **producto punto**
- $\vec{b} \cdot \vec{b} = \|\vec{b}\|^2$ es el cuadrado de la **norma** (longitud) de $\vec{b}$

El **componente escalar de la proyecciÃ³n** (un nÃºmero, no un vector) es:

$$\text{comp}_{\vec{b}}(\vec{a}) = \frac{\vec{a} \cdot \vec{b}}{\|\vec{b}\|}$$

Este valor nos dice **cuÃ¡nto** de $\vec{a}$ va en la direcciÃ³n de $\vec{b}$:
- **Positivo y grande:** $\vec{a}$ apunta mayormente en la misma direcciÃ³n que $\vec{b}$ â†’ **alta similitud**
- **Cercano a cero:** $\vec{a}$ es casi perpendicular a $\vec{b}$ â†’ **poca o ninguna similitud**
- **Negativo:** $\vec{a}$ apunta en direcciÃ³n opuesta a $\vec{b}$ â†’ **caracterÃ­sticas opuestas**

> ğŸ”‘ **ConexiÃ³n con Machine Learning:** En IA, la medida de similitud mÃ¡s usada es la **similitud coseno**:
> $$\cos(\theta) = \frac{\vec{a} \cdot \vec{b}}{\|\vec{a}\| \, \|\vec{b}\|}$$
> Observa que estÃ¡ directamente relacionada con la proyecciÃ³n. Es el componente escalar de la proyecciÃ³n dividido entre $\|\vec{a}\|$. Toma valores entre $-1$ (opuestos) y $1$ (idÃ©ntica direcciÃ³n).

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CELDA DE CONFIGURACIÃ“N â€” Ejecuta esta celda primero (Shift+Enter)
# No necesitas modificar nada aquÃ­. Solo ejecÃºtala.
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from ipywidgets import interact, FloatSlider, Dropdown
from IPython.display import display
import warnings
warnings.filterwarnings('ignore')

# ConfiguraciÃ³n de grÃ¡ficos
plt.rcParams['figure.figsize'] = (8, 8)
plt.rcParams['font.size'] = 12
plt.rcParams['figure.dpi'] = 100

# â”€â”€â”€ Base de datos de pelÃ­culas (cada pelÃ­cula es un vector 2D) â”€â”€â”€
PELICULAS = {
    'Terminator':           np.array([5.0, 1.0]),
    'Titanic':              np.array([1.0, 5.0]),
    'Sr. y Sra. Smith':     np.array([4.0, 4.0]),
    'John Wick':            np.array([5.0, 0.5]),
    'Orgullo y Prejuicio':  np.array([0.5, 5.0]),
    'Matrix':               np.array([4.5, 1.5]),
    'Diario de una PasiÃ³n': np.array([0.5, 4.5]),
    'MisiÃ³n Imposible':     np.array([4.0, 2.0]),
}

# â”€â”€â”€ Colores para las grÃ¡ficas â”€â”€â”€
COL_A    = '#2196F3'   # Azul  â€” vector a
COL_B    = '#F44336'   # Rojo  â€” vector b
COL_PROY = '#4CAF50'   # Verde â€” proyecciÃ³n
COL_PERP = '#FF9800'   # Naranja â€” componente perpendicular

def dibujar_proyeccion(a, b, nombre_a='a', nombre_b='b',
                       titulo='ProyecciÃ³n Ortogonal', mostrar_calculos=True):
    """Dibuja los vectores a, b y la proyecciÃ³n de a sobre b."""

    # â”€â”€ CÃ¡lculos â”€â”€
    dot_ab  = np.dot(a, b)
    dot_bb  = np.dot(b, b)
    norm_a  = np.linalg.norm(a)
    norm_b  = np.linalg.norm(b)

    if dot_bb < 1e-12:
        print("âš ï¸ El vector b es (casi) el vector cero. No se puede proyectar.")
        return

    escalar_proy = dot_ab / dot_bb
    proy         = escalar_proy * b
    comp_escalar = dot_ab / norm_b if norm_b > 1e-12 else 0.0
    perp         = a - proy

    cos_theta = np.clip(dot_ab / (norm_a * norm_b + 1e-12), -1, 1)
    angulo    = np.degrees(np.arccos(cos_theta))

    # â”€â”€ Figura â”€â”€
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.set_facecolor('#FAFAFA')
    ax.grid(True, alpha=0.3, linestyle='--')
    ax.axhline(y=0, color='k', linewidth=0.8)
    ax.axvline(x=0, color='k', linewidth=0.8)

    qargs = dict(angles='xy', scale_units='xy', scale=1, width=0.045, zorder=5)
    ax.quiver(0, 0, b[0], b[1],       color=COL_B,    **qargs)
    ax.quiver(0, 0, a[0], a[1],       color=COL_A,    **qargs)
    ax.quiver(0, 0, proy[0], proy[1], color=COL_PROY, **qargs)

    # Componente perpendicular (lÃ­nea punteada)
    ax.plot([proy[0], a[0]], [proy[1], a[1]],
            color=COL_PERP, linewidth=2, linestyle='--', zorder=4)

    # Etiquetas
    off = 0.25
    ax.annotate(f'{nombre_a} = ({a[0]:.1f}, {a[1]:.1f})',
                xy=(a[0], a[1]), xytext=(a[0]+off, a[1]+off),
                fontsize=10, fontweight='bold', color=COL_A,
                bbox=dict(boxstyle='round,pad=0.3', fc='white', ec=COL_A, alpha=0.9))
    ax.annotate(f'{nombre_b} = ({b[0]:.1f}, {b[1]:.1f})',
                xy=(b[0], b[1]), xytext=(b[0]+off, b[1]-off*2.5),
                fontsize=10, fontweight='bold', color=COL_B,
                bbox=dict(boxstyle='round,pad=0.3', fc='white', ec=COL_B, alpha=0.9))
    ax.annotate(f'proy = ({proy[0]:.2f}, {proy[1]:.2f})',
                xy=(proy[0], proy[1]),
                xytext=(proy[0]-off*5, proy[1]-off*2),
                fontsize=9, color=COL_PROY,
                bbox=dict(boxstyle='round,pad=0.3', fc='white', ec=COL_PROY, alpha=0.9))

    # Leyenda manual
    handles = [
        mpatches.Patch(color=COL_A,    label=f'Vector {nombre_a}'),
        mpatches.Patch(color=COL_B,    label=f'Vector {nombre_b}'),
        mpatches.Patch(color=COL_PROY, label=f'ProyecciÃ³n de {nombre_a} sobre {nombre_b}'),
        mpatches.Patch(color=COL_PERP, label='Componente perpendicular'),
    ]
    ax.legend(handles=handles, loc='upper left', fontsize=9, framealpha=0.9)

    # LÃ­mites
    pts = np.array([[0, 0], a, b, proy])
    margen = 2
    lo = pts.min(axis=0) - margen
    hi = pts.max(axis=0) + margen
    rng = max(hi[0]-lo[0], hi[1]-lo[1])
    center = (lo + hi) / 2
    ax.set_xlim(center[0]-rng/2, center[0]+rng/2)
    ax.set_ylim(center[1]-rng/2, center[1]+rng/2)
    ax.set_aspect('equal')
    ax.set_xlabel('Componente 1  â†’  AcciÃ³n', fontsize=12)
    ax.set_ylabel('Componente 2  â†’  Romance', fontsize=12)
    ax.set_title(titulo, fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()

    # â”€â”€ CÃ¡lculos paso a paso â”€â”€
    if mostrar_calculos:
        print("â•" * 58)
        print("ğŸ“Š  CÃLCULOS DETALLADOS")
        print("â•" * 58)
        print(f"  Vector {nombre_a} = ({a[0]:.1f}, {a[1]:.1f})   â€–{nombre_a}â€– = {norm_a:.4f}")
        print(f"  Vector {nombre_b} = ({b[0]:.1f}, {b[1]:.1f})   â€–{nombre_b}â€– = {norm_b:.4f}")
        print()
        print(f"  Producto punto:  {nombre_a}Â·{nombre_b} = "
              f"({a[0]:.1f})({b[0]:.1f}) + ({a[1]:.1f})({b[1]:.1f}) = {dot_ab:.2f}")
        print(f"  â€–{nombre_b}â€–Â² = {dot_bb:.2f}")
        print()
        print(f"  Escalar de proyecciÃ³n = ({nombre_a}Â·{nombre_b}) / â€–{nombre_b}â€–Â²"
              f" = {dot_ab:.2f} / {dot_bb:.2f} = {escalar_proy:.4f}")
        print(f"  Vector proyecciÃ³n = {escalar_proy:.4f} Ã— ({b[0]:.1f}, {b[1]:.1f})"
              f" = ({proy[0]:.4f}, {proy[1]:.4f})")
        print(f"  â€–ProyecciÃ³nâ€– = {np.linalg.norm(proy):.4f}")
        print()
        print(f"  Componente escalar = ({nombre_a}Â·{nombre_b}) / â€–{nombre_b}â€–"
              f" = {dot_ab:.2f} / {norm_b:.4f} = {comp_escalar:.4f}")
        print(f"  Ãngulo entre {nombre_a} y {nombre_b}: {angulo:.1f}Â°")
        print(f"  Similitud coseno cos(Î¸) = {cos_theta:.4f}")
        print("â•" * 58)

print("âœ… Â¡ConfiguraciÃ³n completada con Ã©xito! ContinÃºa ejecutando las celdas siguientes.")
```

## ğŸ¬ Parte 2: Comparando PelÃ­culas Reales con ProyecciÃ³n

Ahora vamos a aplicar la proyecciÃ³n a nuestro catÃ¡logo de pelÃ­culas. La siguiente herramienta te permite **seleccionar dos pelÃ­culas** de la base de datos y ver:

1. Sus vectores dibujados en el plano AcciÃ³nâ€“Romance.
2. La **proyecciÃ³n** de la primera sobre la segunda.
3. Todos los **cÃ¡lculos paso a paso** (producto punto, normas, componente escalar, Ã¡ngulo, similitud coseno).

### Â¿QuÃ© debes hacer?

1. **Ejecuta la celda de abajo.**
2. **Selecciona diferentes pares** de pelÃ­culas en los menÃºs desplegables.
3. **Compara** los resultados entre pares. FÃ­jate especialmente en la similitud coseno y en el Ã¡ngulo.

> ğŸ’¡ **Sugerencia:** Empieza con pares "obvios" (como **Terminator** y **John Wick**, que ambas son de acciÃ³n) y luego prueba pares "opuestos" (como **Terminator** y **Orgullo y Prejuicio**). DespuÃ©s prueba pares intermedios como **Sr. y Sra. Smith** con otras.

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# HERRAMIENTA 2: Comparador de PelÃ­culas
# âœ Selecciona dos pelÃ­culas en los menÃºs desplegables
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

@interact(
    pelicula_a=Dropdown(options=list(PELICULAS.keys()), value='Terminator',
                        description='PelÃ­cula A:', style={'description_width': '100px'},
                        layout={'width': '350px'}),
    pelicula_b=Dropdown(options=list(PELICULAS.keys()), value='Titanic',
                        description='PelÃ­cula B:', style={'description_width': '100px'},
                        layout={'width': '350px'}),
)
def comparar_peliculas(pelicula_a, pelicula_b):
    a = PELICULAS[pelicula_a]
    b = PELICULAS[pelicula_b]
    dibujar_proyeccion(a, b, nombre_a=pelicula_a, nombre_b=pelicula_b,
                       titulo=f'ComparaciÃ³n: {pelicula_a}  vs  {pelicula_b}')

    # Indicador visual de similitud
    cos_sim = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
    print()
    if cos_sim > 0.9:
        print(f"   â†’ ğŸŸ¢ Â¡Muy similares! (cos = {cos_sim:.3f})  Excelente candidata para recomendaciÃ³n.")
    elif cos_sim > 0.7:
        print(f"   â†’ ğŸŸ¡ Bastante similares (cos = {cos_sim:.3f}).")
    elif cos_sim > 0.4:
        print(f"   â†’ ğŸŸ  Similitud moderada (cos = {cos_sim:.3f}).")
    else:
        print(f"   â†’ ğŸ”´ Muy diferentes (cos = {cos_sim:.3f}). No recomendarÃ­a esta pelÃ­cula.")
```

## ğŸ”„ Parte 3: Â¿CuÃ¡ndo dos pelÃ­culas no tienen nada en comÃºn?

Un caso especial e importante es cuando dos vectores son **ortogonales** (perpendiculares, a 90Â°). En ese caso:
- El producto punto es **cero**: $\vec{a} \cdot \vec{b} = 0$
- La proyecciÃ³n es el **vector cero**: $\text{proy}_{\vec{b}}(\vec{a}) = \vec{0}$
- La similitud coseno es **cero**: $\cos(\theta) = 0$

Esto significa que las dos pelÃ­culas **no comparten ninguna informaciÃ³n en comÃºn**.

En la siguiente herramienta, la pelÃ­cula **Terminator** $(5, 1)$ estÃ¡ fija (vector rojo). TÃº controlas una "PelÃ­cula X" rotando su direcciÃ³n con un slider de Ã¡ngulo. AsÃ­ podrÃ¡s observar paso a paso quÃ© pasa con la proyecciÃ³n cuando el Ã¡ngulo cambia de 0Â° a 180Â°.

### Â¿QuÃ© debes hacer?

1. **Ejecuta la celda de abajo.**
2. **Mueve el slider de Ã¡ngulo lentamente** de 0Â° a 180Â°.
3. Observa cÃ³mo cambia el vector de proyecciÃ³n (verde) y los valores numÃ©ricos.
4. Presta **especial atenciÃ³n** a lo que ocurre en los Ã¡ngulos **0Â°**, **45Â°**, **90Â°**, **135Â°** y **180Â°**.

> ğŸ’¡ **Sugerencia:** Mueve el slider de 5Â° en 5Â° para ver una transiciÃ³n suave. Anota quÃ© ocurre en los Ã¡ngulos clave.

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# HERRAMIENTA 3: Explorador de Ãngulo y Ortogonalidad
# âœ Rota la "PelÃ­cula X" alrededor de Terminator y observa la proyecciÃ³n
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

@interact(
    angulo=FloatSlider(min=0, max=180, step=5, value=30,
                       description='Ãngulo (Â°):',
                       style={'description_width': '100px'},
                       layout={'width': '500px'},
                       readout_format='.0f'),
    magnitud=FloatSlider(min=1, max=5, step=0.5, value=4,
                         description='â€–PelÃ­cula Xâ€–:',
                         style={'description_width': '100px'},
                         layout={'width': '500px'}),
)
def explorador_angulo(angulo, magnitud):
    b = np.array([5.0, 1.0])  # Terminator (FIJO)

    # Crear vector "PelÃ­cula X" rotado respecto a Terminator
    angulo_b = np.arctan2(b[1], b[0])        # Ã¡ngulo de Terminator
    angulo_total = angulo_b + np.radians(angulo)  # rotar desde Terminator
    a = magnitud * np.array([np.cos(angulo_total), np.sin(angulo_total)])

    dibujar_proyeccion(a, b,
                       nombre_a='PelÃ­cula X', nombre_b='Terminator',
                       titulo=f'Ãngulo entre PelÃ­cula X y Terminator: {angulo:.0f}Â°')

    # Indicadores para Ã¡ngulos clave
    print()
    if abs(angulo) < 1:
        print("ğŸ“Œ Ãngulo â‰ˆ 0Â°:   Misma direcciÃ³n â†’ MÃXIMA similitud")
    elif abs(angulo - 45) < 3:
        print("ğŸ“Œ Ãngulo â‰ˆ 45Â°:  DirecciÃ³n intermedia â†’ Similitud moderada-alta")
    elif abs(angulo - 90) < 3:
        print("ğŸ“Œ Ãngulo = 90Â°:  PERPENDICULARES (ortogonales) â†’ Similitud = 0")
        print("   Â¡Este es el punto clave! La proyecciÃ³n es el vector cero.")
    elif abs(angulo - 135) < 3:
        print("ğŸ“Œ Ãngulo â‰ˆ 135Â°: MÃ¡s opuestos que similares â†’ Similitud NEGATIVA")
    elif abs(angulo - 180) < 3:
        print("ğŸ“Œ Ãngulo â‰ˆ 180Â°: Direcciones OPUESTAS â†’ Similitud mÃ­nima (-1)")
```

## ğŸ† Parte 4: Simulando un Sistema de RecomendaciÃ³n Completo

Ahora vamos a unir todo lo que has explorado para simular un **sistema de recomendaciÃ³n real** como el que usarÃ­a Netflix.

### Escenario

Un usuario acaba de ver y disfrutar una pelÃ­cula. Tu sistema debe:
1. **Calcular la proyecciÃ³n** del vector de la pelÃ­cula del usuario sobre cada pelÃ­cula candidata del catÃ¡logo.
2. **Calcular la similitud coseno** con cada candidata.
3. **Ordenar** las candidatas de mÃ¡s similar a menos similar.
4. **Recomendar** las que estÃ¡n en los primeros lugares.

### Â¿QuÃ© verÃ¡s?

- **GrÃ¡fica izquierda:** Todas las pelÃ­culas dibujadas como vectores. La pelÃ­cula del usuario aparece con una estrella â­.
- **GrÃ¡fica derecha:** Un ranking (barras horizontales) ordenado por similitud coseno, de mayor a menor.
- **Debajo:** Los cÃ¡lculos detallados para cada pelÃ­cula.

### Â¿QuÃ© debes hacer?

1. **Ejecuta la celda de abajo.**
2. **Selecciona diferentes pelÃ­culas** del usuario en el menÃº desplegable.
3. **Observa cÃ³mo cambia el ranking** de recomendaciones.
4. **Compara** los rankings de diferentes pelÃ­culas base.

> ğŸ’¡ **Sugerencia:** Prueba seleccionando "Sr. y Sra. Smith" como pelÃ­cula del usuario y observa que sus recomendaciones son mÃ¡s "mezcladas" que las de "Terminator". Piensa por quÃ© ocurre esto (pista: observa la direcciÃ³n de su vector).

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# HERRAMIENTA 4: Sistema de RecomendaciÃ³n de PelÃ­culas
# âœ Selecciona la pelÃ­cula que vio el usuario y mira el ranking
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

@interact(
    pelicula_usuario=Dropdown(
        options=list(PELICULAS.keys()),
        value='Terminator',
        description='PelÃ­cula vista:',
        style={'description_width': '120px'},
        layout={'width': '400px'}
    )
)
def sistema_recomendacion(pelicula_usuario):
    vec_usuario = PELICULAS[pelicula_usuario]

    # Calcular similitud con todas las demÃ¡s pelÃ­culas
    resultados = []
    for nombre, vec in PELICULAS.items():
        if nombre == pelicula_usuario:
            continue
        dot_ab   = np.dot(vec_usuario, vec)
        norm_a   = np.linalg.norm(vec_usuario)
        norm_b   = np.linalg.norm(vec)
        cos_sim  = dot_ab / (norm_a * norm_b)
        comp_esc = dot_ab / norm_b
        angulo   = np.degrees(np.arccos(np.clip(cos_sim, -1, 1)))
        esc_proy = dot_ab / np.dot(vec, vec)
        proy     = esc_proy * vec

        resultados.append({
            'nombre': nombre, 'vector': vec,
            'cos_sim': cos_sim, 'comp_esc': comp_esc,
            'angulo': angulo, 'proy': proy,
        })

    resultados.sort(key=lambda x: x['cos_sim'], reverse=True)

    # â”€â”€ GrÃ¡ficas â”€â”€
    fig, axes = plt.subplots(1, 2, figsize=(16, 7))

    # Izquierda: mapa de vectores
    ax = axes[0]
    ax.set_facecolor('#FAFAFA')
    ax.grid(True, alpha=0.3, linestyle='--')
    ax.axhline(y=0, color='k', linewidth=0.8)
    ax.axvline(x=0, color='k', linewidth=0.8)

    ax.quiver(0, 0, vec_usuario[0], vec_usuario[1],
              angles='xy', scale_units='xy', scale=1,
              color='#E91E63', width=0.07, zorder=10)
    ax.annotate(f'â­ {pelicula_usuario}',
                xy=(vec_usuario[0], vec_usuario[1]),
                xytext=(vec_usuario[0]+0.2, vec_usuario[1]+0.3),
                fontsize=10, fontweight='bold', color='#E91E63')

    colores = plt.cm.tab10(np.linspace(0, 1, len(resultados)))
    for i, res in enumerate(resultados):
        v = res['vector']
        ax.quiver(0, 0, v[0], v[1],
                  angles='xy', scale_units='xy', scale=1,
                  color=colores[i], width=0.04, alpha=0.7, zorder=5)
        ax.annotate(res['nombre'], xy=(v[0], v[1]),
                    xytext=(v[0]+0.15, v[1]+0.15),
                    fontsize=9, color=colores[i])

    ax.set_xlim(-0.5, 6.5)
    ax.set_ylim(-0.5, 6.5)
    ax.set_aspect('equal')
    ax.set_xlabel('AcciÃ³n', fontsize=12)
    ax.set_ylabel('Romance', fontsize=12)
    ax.set_title('Mapa de PelÃ­culas', fontsize=13, fontweight='bold')

    # Derecha: ranking de barras
    ax2 = axes[1]
    nombres     = [r['nombre'] for r in resultados]
    similitudes = [r['cos_sim'] for r in resultados]
    colores_bar = ['#4CAF50' if s > 0.8 else '#FFC107' if s > 0.5 else '#FF5722'
                   for s in similitudes]

    bars = ax2.barh(range(len(nombres)), similitudes,
                    color=colores_bar, edgecolor='white', height=0.6)
    ax2.set_yticks(range(len(nombres)))
    ax2.set_yticklabels(nombres, fontsize=11)
    ax2.set_xlabel('Similitud Coseno', fontsize=12)
    ax2.set_title(f'Ranking de RecomendaciÃ³n\npara fan de "{pelicula_usuario}"',
                  fontsize=13, fontweight='bold')
    ax2.set_xlim(0, 1.15)
    ax2.invert_yaxis()

    medallas = ['ğŸ¥‡', 'ğŸ¥ˆ', 'ğŸ¥‰'] + [''] * 10
    for i, (bar, sim) in enumerate(zip(bars, similitudes)):
        ax2.text(sim + 0.02, i, f'{sim:.3f}', va='center', fontsize=10, fontweight='bold')
        if i < 3:
            ax2.text(-0.07, i, medallas[i], va='center', fontsize=14)

    plt.tight_layout()
    plt.show()

    # â”€â”€ Tabla de resultados â”€â”€
    print("â•" * 65)
    print(f"ğŸ“Š RANKING DE RECOMENDACIÃ“N  â€”  Fan de \"{pelicula_usuario}\"")
    print(f"   Vector del usuario: ({vec_usuario[0]}, {vec_usuario[1]})")
    print("â•" * 65)

    for i, res in enumerate(resultados):
        m = medallas[i] if i < 3 else '  '
        print(f"\n  {m} #{i+1}. {res['nombre']}")
        print(f"      Vector: ({res['vector'][0]}, {res['vector'][1]})")
        print(f"      Componente escalar de proyecciÃ³n: {res['comp_esc']:.4f}")
        print(f"      Similitud coseno: {res['cos_sim']:.4f}")
        print(f"      Ãngulo: {res['angulo']:.1f}Â°")

    print("\n" + "â•" * 65)
```

Si necesitas **verificar algÃºn cÃ¡lculo** manualmente o explorar vectores adicionales, usa esta calculadora. Ingresa las componentes de los vectores $\vec{a}$ y $\vec{b}$ y obtendrÃ¡s todos los cÃ¡lculos paso a paso, **sin grÃ¡fica**, para que puedas concentrarte en los nÃºmeros.

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CALCULADORA: ProyecciÃ³n paso a paso (sin grÃ¡fica)
# âœ Usa los sliders para ingresar cualquier par de vectores
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

@interact(
    a1=FloatSlider(min=-10, max=10, step=0.5, value=5.0,
                   description='aâ‚:', style={'description_width': '40px'},
                   layout={'width': '400px'}),
    a2=FloatSlider(min=-10, max=10, step=0.5, value=1.0,
                   description='aâ‚‚:', style={'description_width': '40px'},
                   layout={'width': '400px'}),
    b1=FloatSlider(min=-10, max=10, step=0.5, value=1.0,
                   description='bâ‚:', style={'description_width': '40px'},
                   layout={'width': '400px'}),
    b2=FloatSlider(min=-10, max=10, step=0.5, value=5.0,
                   description='bâ‚‚:', style={'description_width': '40px'},
                   layout={'width': '400px'}),
)
def calculadora(a1, a2, b1, b2):
    a = np.array([a1, a2])
    b = np.array([b1, b2])

    dot_ab  = np.dot(a, b)
    dot_bb  = np.dot(b, b)
    norm_a  = np.linalg.norm(a)
    norm_b  = np.linalg.norm(b)

    print("â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—")
    print("â•‘           ğŸ§®  CALCULADORA DE PROYECCIÃ“N                â•‘")
    print("â• â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•£")
    print(f"â•‘  Vector a = ({a[0]:.1f}, {a[1]:.1f})")
    print(f"â•‘  Vector b = ({b[0]:.1f}, {b[1]:.1f})")
    print("â• â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•£")
    print(f"â•‘  PASO 1: Producto punto")
    print(f"â•‘    aÂ·b = ({a[0]:.1f})({b[0]:.1f}) + ({a[1]:.1f})({b[1]:.1f})"
          f" = {a[0]*b[0]:.2f} + {a[1]*b[1]:.2f} = {dot_ab:.2f}")
    print(f"â•‘")
    print(f"â•‘  PASO 2: Normas")
    print(f"â•‘    â€–aâ€– = âˆš({a[0]:.1f}Â² + {a[1]:.1f}Â²)"
          f" = âˆš{a[0]**2 + a[1]**2:.2f} = {norm_a:.4f}")
    print(f"â•‘    â€–bâ€– = âˆš({b[0]:.1f}Â² + {b[1]:.1f}Â²)"
          f" = âˆš{b[0]**2 + b[1]**2:.2f} = {norm_b:.4f}")
    print(f"â•‘    â€–bâ€–Â² = {dot_bb:.2f}")
    print(f"â•‘")

    if dot_bb < 1e-12:
        print(f"â•‘  âš ï¸ b es el vector cero. No se puede proyectar.")
    else:
        esc = dot_ab / dot_bb
        proy = esc * b
        comp = dot_ab / norm_b
        cos_t = np.clip(dot_ab / (norm_a * norm_b + 1e-12), -1, 1)
        ang = np.degrees(np.arccos(cos_t))

        print(f"â•‘  PASO 3: Escalar de proyecciÃ³n")
        print(f"â•‘    (aÂ·b)/(bÂ·b) = {dot_ab:.2f} / {dot_bb:.2f} = {esc:.4f}")
        print(f"â•‘")
        print(f"â•‘  PASO 4: Vector proyecciÃ³n")
        print(f"â•‘    proy_b(a) = {esc:.4f} Ã— ({b[0]:.1f}, {b[1]:.1f})"
              f" = ({proy[0]:.4f}, {proy[1]:.4f})")
        print(f"â•‘    â€–proyâ€– = {np.linalg.norm(proy):.4f}")
        print(f"â•‘")
        print(f"â•‘  PASO 5: Componente escalar")
        print(f"â•‘    comp_b(a) = (aÂ·b)/â€–bâ€– = {dot_ab:.2f}/{norm_b:.4f} = {comp:.4f}")
        print(f"â•‘")
        print(f"â•‘  PASO 6: Similitud coseno y Ã¡ngulo")
        print(f"â•‘    cos(Î¸) = (aÂ·b)/(â€–aâ€–Â·â€–bâ€–)"
              f" = {dot_ab:.2f}/({norm_a:.4f}Ã—{norm_b:.4f}) = {cos_t:.4f}")
        print(f"â•‘    Î¸ = {ang:.1f}Â°")

    print("â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•")
```
